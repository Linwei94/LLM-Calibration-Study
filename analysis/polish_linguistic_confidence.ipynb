```python
from bs4 import BeautifulSoup


file_path = "../results/linguistic-judges/mmlu_pro_Meta-Llama-3.1-405B-Instruct-Turbo_verbal_linguistic_shared_sampling_None_gpt-4.1-nano_dec_judge.html"

# 读取 HTML 文件
with open(file_path, "r", encoding="utf-8") as f:
    soup = BeautifulSoup(f, "html.parser")

# 查找所有包含 Sampled message 的块
results = []

for h3 in soup.find_all("h3"):
    if h3.text.strip() == "Sampled message":
        assistant_div = h3.find_next_sibling("div", class_="message assistant")
        results_h3 = assistant_div.find_next_sibling("h3")
        # 确保是对应的 Results 部分
        if results_h3 and results_h3.text.strip() == "Results":
            # 查找 Extracted Answer
            extracted_answer_p = results_h3.find_next_sibling("p", string=lambda t: t and "Extracted Answer:" in t)
            if extracted_answer_p and "None" in extracted_answer_p.text:
                content = assistant_div.find("div", class_="content")
                if content:
                    results.append(content.get_text(strip=True))

# 输出结果
for i, text in enumerate(results, 1):
    print(f"[Block {i} Sampled Message with Extracted Answer: None]\n{text}\n{'-'*80}")
```
