```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
responses = pd.read_csv("paper_hedging_words_results.csv")
```

```python
sorted_order = (
    responses.groupby("hedging_word")["extracted_confidence"]
    .median()
    .sort_values(ascending=True)
    .index
)
```

```python
# Create the boxplot
plt.figure(figsize=(10, 5))
sns.boxplot(
    data=responses,
    x="extracted_confidence",
    y="hedging_word",
    order=sorted_order,
    showfliers=False,  # Hide outliers
    color="lightblue",
    medianprops={"color": "black", "linewidth": 2.5}  # Bold black median line
)

# Customize the plot
plt.xlabel("Perceived Confidence")
plt.ylabel("")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("hedging_words.pdf", dpi=300)
# Show the plot
plt.show()
```
